```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

#### PART 1 - Generate data set *(20 points)*

1. Generate a toy data set with 100 regularly-spaced points and as described by Eq. 8.75-- a simple model where $x$ and $y$ satisfy the following:
$$ 0 \le x_i \le 3 $$
$$ y_i = x_i \sin(x_i) + \epsilon_i $$
where the noise is drawn from a normal distribution $\epsilon_i \sim \rm{N}(0, 0.1)$. *(10 points)*  

2. Plot your toy data set (Refer to Fig. 8.12 for comparison). *(10 points)*


```python
# Generate 100 regularly-spaced points
N = 100
xx = np.linspace(0, 3, N)
np.random.seed(42)

# Generate noise from normal distribution
mu, sigma = 0, 0.2
s = np.random.normal(mu, sigma, 100)

# Generate y values for dataset
yy = xx * np.sin(xx) + s
```

```python
plt.scatter(xx, yy, marker='x')
plt.xlabel("x")
plt.ylabel("y")
```

PART 2 - Apply cross-validation *(80 points)*

Recreate the top panel of Figure 8.14: rms error vs. polynomial degree for the training set and cross-validation set. To do this, you will perform the ff. steps:

1. Split the data set into training, cross-validation, and test sets with 50%, 25% and 25% ratio. You can use sklearn library's model_selection.train_test_split function [(link)](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html). *(20 pts.)*
2. Get the best-fit curve for polynomial degrees $d=0$ to 14 for the training set. You can use numpy library's polyfit function [(link)](https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html). *(30 pts.)*
3. Get the rms errors for both the training and cross-validation sets (for the best-fit curve from Step 2). *(10 pts.)*
4. Plot rms errors for both the training and cross-validation sets against polynomial degree $d$ (as in Fig. 8.14). *(20 pts.)*


```python
from sklearn.model_selection import train_test_split

# Split dataset to training, test, and validation
x_train, xx_test, y_train, yy_test = train_test_split(xx, yy, test_size=0.5, random_state=42)
x_test, x_val, y_test, y_val = train_test_split(xx_test, yy_test, test_size=0.5, random_state=42)
```

```python
def rmse(est, actual):
  return np.sqrt(np.square(np.subtract(actual, est)).mean())    # RMSE formula
```

```python
degrees = np.arange(0, 15, 1)   # Define range of degrees to be tried

training_fit = []   # Store training best fit line
cval_fit = []
cval_rmse = []  # Store rmses
training_rmse = []

for degree in degrees:
  best_fit = np.poly1d(np.polyfit(x_train, y_train, degree))    # Get a best fit line and turn into equation
  training_fit.append(best_fit(x_train))    # Plug x_train values to bf polynomial  to get the best fit line for the training data
  cval_fit.append(best_fit(x_val))    # Plug cval values to bf poly as above to plot later

  # Compute rmse
  training_rmse.append(rmse(training_fit[degree], y_train))
  cval_rmse.append(rmse(cval_fit[degree], y_val))

  print(f"Degree: {degree}, Training RMSE: {training_rmse[degree]}, CrossVal RMSE: {cval_rmse[degree]}")
```

## RMSE and Polynomial Degree

Printed above are the root-mean squared errors of the best fit line against the training and cross-validation sets. We observe a decrease in RMSE for the training data given higher polynomial degrees, but a decrease and subsequent increase in RMSE for the cross-validation set. We can observe this better by plotting as below.

```python
# Plot RMSE vs polynomial degree
plt.plot(degrees, training_rmse, label='Training RMSE')
plt.plot(degrees, cval_rmse, label='CrossVal RMSE', linestyle='--')
plt.title("RMSE vs. Polynomial Degree Plot")
plt.grid()
plt.xlabel("Polynomial Degree")
plt.ylabel("RMSE")
plt.legend()
```

As observed in the printed values, the error continuously increases for the training data given higher degrees. But for the cross-validation set, we observe a lowest point for RMSE given by the degree 3 polynomial before again increasing. This is caused by **overfitting**, which we can again visualize directly as below.

```python
# Plot best fit lines for training data
for degree in degrees:
  xvals, yvals = zip(*sorted(zip(x_train, training_fit[degree])))
  plt.plot(xvals, yvals, label=f"Degree {degree}")    # Plot bfts for original dataset

plt.scatter(xx, yy, marker="x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Training set fit plot")
plt.legend(bbox_to_anchor=(1.0, 1.0), loc="upper left")
plt.grid()
plt.tight_layout()
```

A degree 0 polynomial is just a constant horizontal line, a degree 1 is a straight line with a linear slope, degree 2 will have a parabolic shape, and so on. Thus, we see that the degree 1 polynomial above obviously does not fit the data very well. Meanwhile, the degree 2 polynomial matches the sinusoidal shape of the data allowing for a sudden decrease in RMSE error as seen in the sudden fall in the rather large negative slope in the first RMSE vs polynomial degree figure. It's a bit difficult to see with everything plotted, so let's try graphing them one by one and observe the overfitting.

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].scatter(x_train, y_train, marker="x")   # Plot training data
ax[1].scatter(x_val, y_val, marker="x")   # Plot cross-validation data

for i in range(1, 4):
  ax[0].scatter(x_train, training_fit[i], label=f"Degree {i}", s=10)
  ax[1].scatter(x_val, cval_fit[i], label=f"Degree {i}", s=10)

ax[0].set_title("Training Data vs. Best-Fit Lines")
ax[1].set_title("Cross-Validation Data vs. Best-Fit Lines")
ax[0].legend()
ax[1].legend()
```

In the plot above of the best-fit line against the training and cross-validation data, it is obvious that the degree 2 polynomial underfits while the degree 3 polynomial follows the shape of the data best for both datasets. Let's check for higher degrees.

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].scatter(x_train, y_train, marker="x")   # Plot training data
ax[1].scatter(x_val, y_val, marker="x")   # Plot cross-validation data

for i in range(10, 13):
  ax[0].scatter(x_train, training_fit[i], label=f"Degree {i}", s=10)
  ax[1].scatter(x_val, cval_fit[i], label=f"Degree {i}", s=10)

ax[0].set_title("Training Data vs. Best-Fit Lines")
ax[1].set_title("Cross-Validation Data vs. Best-Fit Lines")
ax[0].legend()
ax[1].legend()
```

The overfitting is already very observable for these degree values, especially when looking at the left tail of the plot, notice how the best-fit lines of these degrees follow the training data closely along the noise. This causes the best-fit lines to not generalize as well in the cross-validation plot, as it now follows a different shape. This is also the reason why the cross-validation RMSE is lowest for around degree 3 to 5 polynomial. The sinusoidal shape of the original data source is best approximated with these degrees having the same number of "turning points". Any higher polynomial degree and the turns will simply tend to follow noise. Meanwhile, observe below how degrees 3-5 follow the shape of the dataset quite well.

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].scatter(x_train, y_train, marker="x")   # Plot training data
ax[1].scatter(x_val, y_val, marker="x")   # Plot cross-validation data

for i in range(3, 6):
  ax[0].scatter(x_train, training_fit[i], label=f"Degree {i}", s=10)
  ax[1].scatter(x_val, cval_fit[i], label=f"Degree {i}", s=10)

ax[0].set_title("Training Data vs. Best-Fit Lines")
ax[1].set_title("Cross-Validation Data vs. Best-Fit Lines")
ax[0].legend()
ax[1].legend()
```

```python
# Generate 100 regularly-spaced points
N = 100
xx = np.linspace(-3, 3, N)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Create a degree 3 to 5 polynomial
poly3 = xx**3
ax[0].plot(xx, poly3)
ax[0].set_title("Shape of a degree 3 polynomial")

ax[1].scatter(xx, yy, s=2)
ax[1].set_title("Data")
```

Note the number of "main turns" we want to make for a line to fit the data. Thus we expect that the degree 3 (to 5) polynomial generalizes best even in other datasets, even in other seeds considering the original shape of the distribution. We can verify this again using other seeds.
