#### PART 1 - Network generation and characteristics *(30 points)*

1. Using the Python package networkx, define a network that represents the social network of friends in the diagram. *(15 points)*  

2. Calculate the ff. network characteristics: *(15 points; 5 points each)*
- Diameter
- Density
- Average Degree

3. Create the degree distribution plot and briefly discuss what it tells us about this social network. *(10 points)*

```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
G = nx.Graph()

nodes_to_add = ["Jill", "Mike", "Bob", "Emma", "John", "Leah", "Shane",
                "Liz", "Allen", "Lisa"]

G.add_nodes_from(nodes_to_add)

edges_to_add = [("Jill", "Mike"), ("Jill", "Bob"), ("Jill", "John"),
                ("Jill", "Leah"), ("Jill", "Shane"), ("Jill", "Emma"),
                ("Bob", "Emma"), ("Bob", "Mike"), ("Bob", "John"), 
                ("John", "Shane"), ("Leah", "Shane"), ("Leah", "John"),
                ("Emma", "Liz"), ("Emma", "Mike"), ("Emma", "Shane"),
                ("Shane", "Liz"), ("Allen", "Liz"), ("Allen", "Lisa")]

G.add_edges_from(edges_to_add)

nx.draw(G, with_labels=True, node_size=2000)
```

```python
print("Diameter: ", nx.diameter(G))
print("Density : ", nx.density(G))

degrees = [G.degree(n) for n in G.nodes]
avg_degree = np.array(degrees).mean()

print("Average degree: ", avg_degree)
```

The network diameter refers to the shortest distance possible when connecting the two most distant nodes. In this case, the network is relatively short and simple, so we can calculate this by counting. In particular, it is note difficult to see that the most distant nodes will be from Lisa to anyone in the connected subset of the network around Jill. All of them have a shortest distance of $4$ to Lisa.

Network density meanwhile refers to the ratio of actual connections to all potential connections between nodes. For unidirectional nodes, the number of edges is the number of actual connections. Meanwhile the number of potential connections will be given by the number of nodes multiplied to the number of all other nodes divided by two accounting for unidirectionality. This is written simply in the formula,

$$ c_{pot} = n(n - 1)/2, $$
$$ c_{act} = m, $$

where $n$ is the number of nodes and $m$ is the number of edges. In our simple network $m$ is easily counted as $18$ while plugging numbers in the formula for potential connections we obtain $45$. Thus the density is given by,

$$ d = 2m/n(n-1) = 0.4,$$

the same result given above.

Finally, the average degree is simply the ratio of edges to nodes in the network. Thus for the one above, we can manually calculate the average degree as

$$ <deg> = (18*2)/10 = 3.6, $$

where we multiplied the number of edges by two to account for unidirectionality.

```python
plt.hist(degrees, align='mid')
plt.xlabel("Degree")

# or

nx.degree_histogram(G)
```

The degree distribution plot above suggests that most nodes are fairly well connected, with only two nodes having one and two degrees each. Most nodes reside on the well-connected part with Jill on the center. But even from this distribution alone, we can already get an idea of what the network looks like. For example, the one and two degree nodes already suggest an isolated tail-end of the network, with a webbed part when noting the single well-connected node and fairly even degree distribution among the rest.

#### PART 2 - Centrality Measures *(30 points)*

Generate a table of centrality measures per node. Columns are:
- Name
- Degree Centrality
- Betweenness Centrality
- Closeness Centrality

Briefly discuss who are the most "central" people in the social network and why.

```python
dc = nx.degree_centrality(G)
bc = nx.betweenness_centrality(G, normalized=False)
cc = nx.closeness_centrality(G, wf_improved=False)

# Merge as single dictionary
dicts = [dc, bc, cc]
d = {}
for key in dc.keys():
    d[key] = [d[key] for d in dicts]

# Create pandas dataframe
df = pd.DataFrame(d)
df.index = ["Degree Centrality", "Betweeness Centrality", "Closeness Centrality"]
df = df.transpose()
df
```

Degree centrality counts how many edges a node is connected to. That is, the node with the highest degree centrality has the most direct connections. It is easy to identify in the network above that the one with the highest degree centrality is Jill. Likewise, the one with least connections is Lisa linked only to Allen on the tail-end of the network. Note that the degree centrality calculated by networkx is normalized.

Betweeness centrality meanwhile counts the number of shortest path a node is on. That is, a node with a high betweeness centrality means it is important for efficient flow of information. In the network above, it is again not difficult to see how Liz should have the highest betweeness centrality considering all nodes around Jill must pass through her to reach Allen and Lisa. Emma, Shane, and Allen share the same high betweeness centrality for the same reason, but with the limitation that Allen is only on the path for Lisa, while Emma and Shane routes a different set of nodes to the tail of the network. All of them however must pass Liz if the nodes intend to reach Allen and Lisa. Lisa, Mike, and Leah meanwhile don't have friends that they are uniquely connected to, resulting in a betweeness centrality of zero.

Finally, closeness centrality measures how close a node is to all other nodes. Perhaps unsurprisingly, Emma and Shane are highest in terms of closeness centrality with Jill and Liz coming second due to their literal "central" location in the network. Likewise, the farthest to all other nodes is Lisa followed by Allen in the tail-end of the network, resulting in their low closeness centrality values.

#### PART 3 - Identify communities *(40 points)*

1. Identify communities in the social network using *(30 points)*:
- Louvain community detection algorithm
- Any other community detection algorithm of your choice

2. Calculate the modularity of the partitions from the two community detection algorithms. *(10 points)*

Briefly describe the algorithm you used and compare the results from the two algorithms.

```python
louvain = nx.community.louvain_communities(G, seed=42)

louvain
```

```python
nx.community.modularity(G, louvain)
```

```python
node_colors = []
color_map = ['r', 'g','b']

for node in G.nodes():
  community_found = False
  for i, community in enumerate(louvain):
    if node in community:
      node_colors.append(color_map[i % len(color_map)])  # Use modulo to handle more communities than colors
      community_found = True
      break  # Exit inner loop once community is found for the node

pos = nx.spring_layout(G, seed=100)  # Get node positions 
plt.title('Louvain communities')
nx.draw(G, pos=pos, node_color=node_colors, node_size=2000, font_color='white', with_labels=True)
```

The modularity of a community, as the name suggests, is the measure of how densely connected a set of nodes is in a given community. That is, given a network split into communities (or modules), high modularity means that the subset of the network is well-connected with nodes within the same community, while being less connected with those in another. The louvain community detection algorithm attempts to optimize modularity, and as we see in the colored nodes above, it indeed seems to have partitioned the nodes to maximize the density of connections of the communities. In this case we obtain a modularity value of $0.2222$. We can compare this with another community detection algorithm below.

```python
greedy = nx.community.greedy_modularity_communities(G)

node_colors = []
color_map = ['r', 'g','b']

for node in G.nodes():
  community_found = False
  for i, community in enumerate(greedy):
    if node in community:
      node_colors.append(color_map[i % len(color_map)])  # Use modulo to handle more communities than colors
      community_found = True
      break  # Exit inner loop once community is found for the node

pos = nx.spring_layout(G, seed=100)  # Get node positions 
plt.title('Greedy communities')
nx.draw(G, pos=pos, node_color=node_colors, node_size=2000, font_color='white', with_labels=True)

nx.community.modularity(G, greedy)
```

The greedy modularity algorithm as the name suggests seeks partition the network to communities based on maximum modularity. In this case, we observe that it partitioned the community in the same way and achieved the same modularity value as in the Louvain algorithm, which is expected considering their metrics for partitioning.

```python
kern = nx.community.kernighan_lin_bisection(G, partition=None, max_iter=10, weight='weight', seed=None)

node_colors = []
color_map = ['r', 'g','b']

for node in G.nodes():
  community_found = False
  for i, community in enumerate(kern):
    if node in community:
      node_colors.append(color_map[i % len(color_map)])  # Use modulo to handle more communities than colors
      community_found = True
      break  # Exit inner loop once community is found for the node

pos = nx.spring_layout(G, seed=100)  # Get node positions 
plt.title('Kernighan-Lin communities')
nx.draw(G, pos=pos, node_color=node_colors, node_size=2000, font_color='white', with_labels=True)

nx.community.modularity(G, kern)
```

Above we also attempt to use the Kernighan-Lin algorithm. In this case, the algorithm seeks to partition the network into two while reducing edge cuts between the resulting communities. The partition above indeed splits the community into two, but there is not a unique result to the algorithm as we see below by running the code again.

```python
kern = nx.community.kernighan_lin_bisection(G, partition=None, max_iter=10, weight='weight', seed=None)

node_colors = []
color_map = ['r', 'g','b']

for node in G.nodes():
  community_found = False
  for i, community in enumerate(kern):
    if node in community:
      node_colors.append(color_map[i % len(color_map)])  # Use modulo to handle more communities than colors
      community_found = True
      break  # Exit inner loop once community is found for the node

pos = nx.spring_layout(G, seed=100)  # Get node positions 
plt.title('Kernighan-Lin communities')
nx.draw(G, pos=pos, node_color=node_colors, node_size=2000, font_color='white', with_labels=True)

nx.community.modularity(G, kern)
```

```python
kern = nx.community.kernighan_lin_bisection(G, partition=None, max_iter=10, weight='weight', seed=None)

node_colors = []
color_map = ['r', 'g','b']

for node in G.nodes():
  community_found = False
  for i, community in enumerate(kern):
    if node in community:
      node_colors.append(color_map[i % len(color_map)])  # Use modulo to handle more communities than colors
      community_found = True
      break  # Exit inner loop once community is found for the node

pos = nx.spring_layout(G, seed=100)  # Get node positions 
plt.title('Kernighan-Lin communities')
nx.draw(G, pos=pos, node_color=node_colors, node_size=2000, font_color='white', with_labels=True)

nx.community.modularity(G, kern)
```

The algorithm initially randomly splits the network nodes into two subsets then iterates over each and swapping pairs until there is a minimum number of edges between nodes not of the same community. Since there are multiple ways to arrange the network we have into such, we obtain multiple results depending on the random initial split. Furthermore sincce the algorithm does not mind the modularity value, the modularity scores vary as we saw above.
